```python
#read txt
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
openfile = open('J04.txt','r')
list = []
list = openfile.readlines()

for i in range(len(list)):
    list[i] = list[i].replace('\n','')
    
#extract number from list   
list1 = []
for i in range(len(list)):
    list1.append(re.findall(r'\d+',list[i]))
list = list1

#list from str to int
for i in range(len(list)):
    for j in range(len(list[i])):
        list[i][j] = int(list[i][j])
```

```python
tot = 0
for i in range(len(list)):
    if list[i][0] <= list[i][2] and list[i][1] >= list[i][3]:
        tot += 1
        print("1:",list[i])
    elif list[i][2] <= list[i][0] and list[i][3] >= list[i][1]:
        tot += 1
        print("2:",list[i])
print(tot)
```
